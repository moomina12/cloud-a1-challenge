```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from selenium.webdriver.common.action_chains import ActionChains
from time import sleep
import random
import pandas as pd
```

```python
driver=webdriver.Chrome()
```

```python
url=driver.get("https://www.indeed.com/")
```

```python
where_input=driver.find_element(By.ID,'text-input-where')
where_input=driver.find_element(By.CLASS_NAME,"yosegi-InlineWhatWhere-primaryButton")
where_input.send_keys("")
where_input.click()
```

```python
find_jobs_button=driver.find_element(By.CLASS_NAME,"yosegi-InlineWhatWhere-primaryButton")
find_jobs_button.click()
```

```python
url = "https://ae.indeed.com/jobs?q=python+developer&l=Dubai&start="
details = []# an empyt list has been created to store all links
for i in range(17):# loop will iterate over the pages and it will load upto 17 pages
    print(f'scrapping page number :{i+1}')
    link = url + str(i+10)
    print(f'link : {link}')
    driver.get(link)
    time.sleep(2)
    for j in driver.find_elements(By.CLASS_NAME,'resultContent'):#second loop will run to find the elements from class "slider_text"
        details.append(j.text.split("\n"))# it will append all data extracted using .text function         
    print('-'*50)
```

```python
job_title=[]
company=[]
salary=[]
for i in details:
    try:
        job_title.append(i[0])
    except:
        job_title.append("")
    try:
        company.append(i[1])
    except:
        company.append("")
    try:
        salary.append(i[3])
    except:
        salary.append("")
    
```

```python
job_title,company, salary
```

```python
data=pd.DataFrame({"Job_Titele":job_title,"Company":company,"Salary":salary})
```

```python
data.to_csv("Python_Developer_Salary-Indeed2.csv")
```
